```python

import numpy as np

def transformar_dataframe(df):
    # Mostrar la forma del DataFrame
    print("Forma inicial del DataFrame:", df.shape)
    
    # Eliminar filas con valores NaN
    df = df.dropna()
    print("Forma del DataFrame después de eliminar NaN:", df.shape)
    
    # Estandarizar los nombres de las cabeceras
    df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace(r'[^a-zA-Z0-9_]', '')
    
    # Identificar y mostrar las columnas numéricas y categóricas
    columnas_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
    columnas_categoricas = df.select_dtypes(exclude=[np.number]).columns.tolist()
    print("Columnas numéricas:", columnas_numericas)
    print("Columnas categóricas:", columnas_categoricas)
    
    # Convertir la columna 'effective_to_date' al formato datetime, extraer el mes
    if 'effective_to_date' in df.columns:
        df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])
        df['mes'] = df['effective_to_date'].dt.month
        # Filtrar por el primer trimestre
        df = df[df['mes'].isin([1, 2, 3])]
    else:
        print("La columna 'effective_to_date' no existe en el DataFrame.")
    
    # Retornar el DataFrame transformado
    return df

# Supongamos que tienes un DataFrame llamado df para transformar
# df_transformado = transformar_dataframe(df)
# print("Forma final del DataFrame:", df_transformado.shape)
```

```python
import matplotlib.pyplot as plt

response_counts = df['response'].value_counts()
response_counts.plot(kind='bar')
plt.title('Número Total de Respuestas')
plt.xlabel('Respuesta')
plt.ylabel('Total')
plt.show()
```

```python
response_rate_by_sales_channel = df.groupby('sales_channel')['response'].value_counts(normalize=True).unstack().fillna(0)['Yes']
response_rate_by_sales_channel.plot(kind='bar')
plt.title('Tasa de Respuesta por Canal de Ventas')
plt.xlabel('Canal de Ventas')
plt.ylabel('Tasa de Respuesta')
plt.show()
```

```python
# tasa de respuesta por total claim amount
bins = pd.cut(df['total_claim_amount'], bins=[0, 200, 400, 600, 800, 1000, 1200, 1400])
response_rate_by_claim_amount = df.groupby(bins)['response'].value_counts(normalize=True).unstack().fillna(0)['Yes']
response_rate_by_claim_amount.plot(kind='bar')
plt.title('Tasa de Respuesta por Total Claim Amount')
plt.xlabel('Total Claim Amount')
plt.ylabel('Tasa de Respuesta')
plt.show()
```

```python
# Tasa de respuesta por ingresos
bins = pd.cut(df['income'], bins=[0, 20000, 40000, 60000, 80000, 100000])
response_rate_by_income = df.groupby(bins)['response'].value_counts(normalize=True).unstack().fillna(0)['Yes']
response_rate_by_income.plot(kind='bar')
plt.title('Tasa de Respuesta por Ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Tasa de Respuesta')
plt.show()
```
